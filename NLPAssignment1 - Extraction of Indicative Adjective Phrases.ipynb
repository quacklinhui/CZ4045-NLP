```python
import nltk
import pandas as pd
import json
import numpy as np
```

```python
# opening dataset
with open('reviewSelected100.json') as f:
    data = json.loads("[" + 
        f.read().replace("}\n{", "},\n{") + 
    "]")
```

```python
# Getting all of the reviews
reviewData = pd.DataFrame(data)
reviewData = reviewData[['business_id', 'text']]
reviewData
```

```python
# grouping dataset by business_id
reviewsGrouped = reviewData.groupby(['business_id'], as_index = False).agg(lambda x: x.tolist())
reviewsGrouped['tokens'] = None
reviewsGrouped['postag'] = None
reviewsGrouped['adj phrases'] = None
reviewsGrouped
```

```python
for i in range(len(reviewsGrouped)):
    for j in range(len(reviewsGrouped['text'][i])):
        tokens = []
        token = nltk.word_tokenize(reviewsGrouped['text'][i][j])
        tokens.append(token)
        reviewsGrouped['tokens'][i] = [reviewsGrouped['tokens'][i]] + tokens
        tokens.clear()
```

```python
for i in range(len(reviewsGrouped)):
    reviewsGrouped['tokens'][i].pop(0)
```

```python
for i in range(len(reviewsGrouped)):
    for j in range(len(reviewsGrouped['tokens'][i])):
        postags = []
        postag = nltk.pos_tag(reviewsGrouped['tokens'][i][j])
        postags.append(postag)
        reviewsGrouped['postag'][i] = [reviewsGrouped['postag'][i]] + postags
        postags.clear()
```

```python
for i in range(len(reviewsGrouped)):
    reviewsGrouped['postag'][i].pop(0)
```

```python
# getting all the adj phrases
for i in range(len(reviewsGrouped)):
    for j in range(len(reviewsGrouped['postag'][i])):
        tree = chunkParser.parse(reviewsGrouped['postag'][i][j])
        adj_phrase = []
        for subtree in tree.subtrees():
            if isinstance(subtree, nltk.tree.Tree):
                if subtree.label() == 'ADJP':
                    for word, tag in subtree.leaves():
                        if (len(subtree.leaves())) > 1:
                            nested_adj_phrase = []
                            for x in range(len(subtree.leaves())):
                                nested_word, nested_tag = (subtree.leaves())[x]
                                nested_adj_phrase.append(nested_word)
                            if nested_adj_phrase not in adj_phrase:    
                                adj_phrase.append(nested_adj_phrase)
                        else:
                            adj_phrase.append(word)
        reviewsGrouped['adj phrases'][i] = [reviewsGrouped['adj phrases'][i]] + adj_phrase      

for i in range(len(reviewsGrouped)):
    reviewsGrouped['adj phrases'][i].pop(0)

# converting all nested lists into str phrases   
for i in range(len(reviewsGrouped)):
    for j in range(len(reviewsGrouped['adj phrases'][i])):
        if type(reviewsGrouped['adj phrases'][i][j]) == list:
            combined_phrase = ''
            for x in range(len(reviewsGrouped['adj phrases'][i][j])):
                combined_phrase = combined_phrase + reviewsGrouped['adj phrases'][i][j][x] + ' '
            reviewsGrouped['adj phrases'][i].append(combined_phrase)
    
    # todo: change code to do until all lists are converted to strings
    # remove all nested lists
    for x in reviewsGrouped['adj phrases'][i]:
        if type(x) == list: 
            reviewsGrouped['adj phrases'][i].remove(x)
    for x in reviewsGrouped['adj phrases'][i]:
        if type(x) == list: 
            reviewsGrouped['adj phrases'][i].remove(x)
    for x in reviewsGrouped['adj phrases'][i]:
        if type(x) == list: 
            reviewsGrouped['adj phrases'][i].remove(x)
```

```python
# getting unique words
uniqueWords = set()
for i in range(len(reviewsGrouped['adj phrases'])):
    uniqueWords = uniqueWords.union(set(reviewsGrouped['adj phrases'][i]))

uniqueWords
```

```python
# Getting dictionary of unique phrases and find number of occurrences of phrase of each businessId 
tfidf_data = reviewsGrouped[['business_id', 'adj phrases']]
tfidf_data['numOfWords'] = None
tfidf_data['Term Frequency'] = None
tfidf_data['TFIDF Score'] = None

for i in range(len(tfidf_data)):
    tfidf_data['numOfWords'][i] = dict.fromkeys(uniqueWords, 0)
    
    for word in tfidf_data['adj phrases'][i]:
        tfidf_data['numOfWords'][i][word] += 1
        
```

```python
# to compute term frequency - the number of times a word appears in a document divided by the total number of words in the document
# in this case, we used number of times a adjective phrase appears in a document divided by the total number of adjective phrases in the document
def computeTF(wordDict, bagOfWords, hasAdjPhrases):
    tfDict = {}
    bagOfWordsCount = len(bagOfWords)
    for word, count in wordDict.items():
        if hasAdjPhrases == True:
            tfDict[word] = count / float(bagOfWordsCount)
        else:
            tfDict[word] = count / 1
    return tfDict
```

```python
# getting phrase frequency for each business
for i in range(len(tfidf_data)):
    if len(tfidf_data['adj phrases'][i]) > 0:
        tfidf_data['Term Frequency'][i] = computeTF(tfidf_data['numOfWords'][i], tfidf_data['adj phrases'][i], True)
    else: 
        tfidf_data['Term Frequency'][i] = computeTF(tfidf_data['numOfWords'][i], tfidf_data['adj phrases'][i], False)
tfidf_data
```

```python
# computing Inverse Data Frequency (IDF) - log of the number of documents divided by the number of documents that contain the word w
def computeIDF(documents):
    import math
    N = len(documents)
    
    idfDict = dict.fromkeys(documents[0].keys(), 0)
    for document in documents:
        for word, val in document.items():
            if val > 0:
                idfDict[word] += 1
                
    for word, val in idfDict.items():
        idfDict[word] = math.log(N/float(val))
        
    return idfDict
```

```python
idfs = computeIDF(tfidf_data['numOfWords'])
idfs
```

```python
# computing TFIDF - TF x IDF
def computeTFIDF(tfBagOfWords, idfs):
    tfidf = {}
    for word, val in tfBagOfWords.items():
        tfidf[word] = val * idfs[word]
        
    return tfidf
```

```python
# computing TFIDF Score, and sorting them in descending order
for i in range(len(tfidf_data)):
    tfidf_data['TFIDF Score'][i] = computeTFIDF(tfidf_data['Term Frequency'][i], idfs)
    tfidf_data['TFIDF Score'][i] = sorted(tfidf_data['TFIDF Score'][i].items(), key=lambda x: x[1], reverse=True)
tfidf_df = pd.DataFrame(tfidf_data[['business_id', 'TFIDF Score']])
tfidf_df
```
